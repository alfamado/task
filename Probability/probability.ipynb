```python
order_payment_review["order_purchase_timestamp_x"] = pd.to_datetime(order_payment_review["order_purchase_timestamp_x"])
order_payment_review["order_delivered_customer_date_x"] = pd.to_datetime(order_payment_review["order_delivered_customer_date_x"])

order_payment_review["delivery_time_days"] = (order_payment_review["order_delivered_customer_date_x"] - order_payment_review["order_purchase_timestamp_x"]).dt.days
delivery_time = order_payment_review["delivery_time_days"].dropna()

mean_delivery = delivery_time.mean()
variance_delivery = delivery_time.var()
std_delivery = delivery_time.std()

print("Expected delivery time (Mean):", round(mean_delivery, 2), "days")
print("Variance of delivery time:", round(variance_delivery, 2))
print("Standard Deviation of delivery time:", round(std_delivery, 2), "days")
```

```python
population_payment_mean = order_payment_review["payment_value_y"].mean()

sample_50 = order_payment_review["payment_value_y"].sample(50, random_state=42)
sample_mean_50 = sample_50.mean()

sample_200 = order_payment_review["payment_value_y"].sample(200, random_state=42)
sample_mean_200 = sample_200.mean()

print("Population mean payment value:", round(population_payment_mean, 2))
print("Sample mean (n=50):", round(sample_mean_50, 2))
print("Sample mean (n=200):", round(sample_mean_200, 2))

```

```python
# Law of Large Number (LLN)
# Draw increasing sample sizes and trzck their means
sample_sizes = [10, 50, 100, 500, 1000, 5000]
sample_mean = []

for n in sample_sizes:
    sample = order_payment_review["payment_value_y"].sample(n, random_state=234)
    sample_mean.append(sample.mean())

for n, mean in zip(sample_sizes, sample_mean):
    print(f"Sample size={n}, sample meamn={round(mean,2)}")

print("Population mean (true):", round(population_payment_mean, 2))
```

```python
review_score = order_reviews["review_score"].dropna()

neutral_score = 3

t_stat, p_value = stats.ttest_1samp(review_score, popmean=neutral_score)

print("Mean review score:", round(review_score.mean(), 2))
print("T-statistic:", round(t_stat, 3))
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject H0: The average review score is significantly different from 3 (neutral).")
else:
    print("Fail to reject H0: The data does not provide enough evidence that the average review differs from 3.")
    
```

```python
# Two-Sample t-test
credit_card = order_payment_review.loc[order_payment_review["payment_type_y"] == "credit_card", "payment_value_y"]
boleto = order_payment_review.loc[order_payment_review["payment_type_y"] == "boleto", "payment_value_y"]

# Dropping missing values just in case
credit_card = credit_card.dropna()
boleto = boleto.dropna()



print("Credit Card Average Spend:", round(credit_card.mean(), 2))
print("Boleto Average Spend:", round(boleto.mean(), 2))
```

```python
# Let perform out t-test
t_stat, p_value = stats.ttest_ind(credit_card, boleto)

print("T-statistic:", t_stat)
print("P-value", p_value)
```
