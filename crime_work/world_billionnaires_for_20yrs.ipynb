```python
# lets do somwe cleaning, here we are cleaning the networth column, we want to remove the dollar symbol and billion from the amount then convert it to float as a variable type

df['Net_worth(USD)'] = df['Net_worth(USD)'].str.replace('$', '', regex=False).str.replace('billion','', regex = False).astype(float)
```
