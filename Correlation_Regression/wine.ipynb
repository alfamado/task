```python
# Set visual style
sns.set(style="whitegrid")
# 1. Correlation Heatmap
plt.figure(figsize=(12, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Wine Quality Features", fontsize=16)
plt.show()
# 2. Correlation of each feature with quality (bar chart)
corr_with_quality = corr["quality"].drop("quality").sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=corr_with_quality.values, y=corr_with_quality.index, palette="viridis")
plt.title("Correlation of Each Feature with Wine Quality", fontsize=16)
plt.xlabel("Correlation Coefficient")
plt.ylabel("Feature")
plt.show()
# 3. Pairplot of top correlated features with quality
top_features = corr_with_quality.head(4).index.tolist()
sns.pairplot(df, vars=top_features + ["quality"], hue="quality", palette="coolwarm", diag_kind="kde")
plt.suptitle("Pairplot of Top Correlated Features with Quality", y=1.02, fontsize=16)
plt.show()
# 4. Boxplots of top correlated features vs quality
plt.figure(figsize=(16,10))
for i, feature in enumerate(top_features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x="quality", y=feature, data=df, palette="coolwarm")
    plt.title(f"{feature.capitalize()} vs Wine Quality")
plt.tight_layout()
plt.show()
```
