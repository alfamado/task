```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

workspace = Path("workspace")
workspace.mkdir(exist_ok=True)
sales_analysis = workspace / "eletronic_sales.xlsx"
```

```python
df = pd.read_excel(sales_analysis)
df.head()
```

```python
df["Revenue"] = round(df["Units"]*df["Price"] * 1000,2)
```

```python
df["Week"] = df["Date"].dt.day_name()
df["Month"] = df["Date"].dt.month_name()
df["Year"] = df["Date"].dt.year
```

```python
df.head()
```

```python
total_product_unit_sold = df["Units"].sum()
print(f"Total Product of units sold: {total_product_unit_sold}")
```

```python
import warnings 
warnings.filterwarnings('ignore')

pd.set_option("display.max_columns", None)
pd.set_option('display.width', 1000)

df
```

```python
Total_Revenue = df["Revenue"].sum()
print(f"Total Revenue Generated: {Total_Revenue}")
```

```python
Average_revenue_per_unit = round(Total_Revenue / total_product_unit_sold,2)
print(f"Average Revenue per unit: {Average_revenue_per_unit}")
```

```python
Number_of_products = df["Products"].nunique()
print(f"Number of Products: {Number_of_products}")
```

```python
df.set_index("Date", inplace=True)
```

```python
sales_for_year_2015 = df[df["Year"] == 2015]
sales_for_year_2015.head(50)
```

```python
sales_for_year_2015.count()
```

```python
Total_monthly_revenue_2015 = sales_for_year_2015.groupby("Month")["Revenue"].sum().sort_index()
print("Total Monthly Revenue is")
Total_monthly_revenue_2015
```

```python
Total_monthly_revenue_sort_2015 = Total_monthly_revenue_2015.sort_values(ascending=True)
print("Total Monthly Revenue sort by amount generated")
Total_monthly_revenue_sort_2015
```

```python
sales_for_year_2014 = df[df["Year"] == 2014]
sales_for_year_2014.head(50)
```

```python
sales_for_year_2014.count()
```

```python
Total_monthly_revenue_2014 = sales_for_year_2014.groupby("Month")["Revenue"].sum().sort_index()
print("Total Monthly Revenue is")
Total_monthly_revenue_2014
```

```python
Total_monthly_revenue_sort_2014 = Total_monthly_revenue_2014.sort_values(ascending=True)
print("Total Monthly Revenue sort by amount generated")
Total_monthly_revenue_sort_2014
```

```python
Total_yearly_revenue = df.groupby("Year")["Revenue"].sum().sort_index()
print("Total Yearly Revenue is")
Total_yearly_revenue
```
