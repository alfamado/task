```python
age = df["age"]

age_mean = age.mean()
age_median = age.median()
age_mode = age.mode()[0]
age_skew = age.skew()
age_min = age.min()
age_max = age.max()
age_range = age_max - age_min
age_variance = age.var()
age_stddev = age.std()
age_percentile = age.quantile([0.10, 0.25, 0.50, 0.75])
age_IQR = age_percentile[0.75] - age_percentile[0.25]
age_outlier_lower = age_percentile[0.25] - (age_IQR * 1.5)
age_outlier_upper = age_percentile[0.75] + (age_IQR * 1.5)
age_coefficient_variation = (age_stddev / age_mean) * 100
```

```python
print(age_mean)
print(age_median)
print(age_mode)
print(age_skew)
print(f"age minimum: {age_min}")
print(f"age maximum: {age_max}")
print(f"age range: {age_range}")
print(f"age Variance: {age_variance}")
print(f"age Standard Deviation: {age_stddev}")
print(f"age percentile:\n{age_percentile}")
print(f"age Inter Quartile Range: {age_IQR}")
print(f"age Outlier Lower: {age_outlier_lower}")
print(f"age Outlier Upper: {age_outlier_upper}")
print(f"belly_coefficient_variation: {age_coefficient_variation}")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(age, bins=30, kde=True, color="purple")

# Add vertical lines
plt.axvline(age_mean, color='green', linestyle='--', linewidth=2, label=f'Mean = {age_mean:.2f}')
plt.axvline(age_median, color='grey', linestyle='-', linewidth=2, label=f'Median = {age_median:.2f}')
plt.axvline(age_mode, color='red', linestyle=':', linewidth=2, label=f'Mode = {age_mode:.2f}')

# Add title and skewness
plt.title(f"Age Description (Skewness = {age_skew:.2f})", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```

```python
kurtosis = df['age'].kurtosis()
print(kurtosis)
```
